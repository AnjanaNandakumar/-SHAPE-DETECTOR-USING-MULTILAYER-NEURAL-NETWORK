```python
#libraries_imported 
import cv2
import numpy as np
import os         
from random import shuffle 
from tqdm import tqdm      
import tensorflow as tf
import matplotlib.pyplot as plt
import tflearn
from keras.optimizers import SGD
```

```python
#data cleaning and data shuffling 
triangles = []

# Load in the images
for filepath in os.listdir(r'D:/sherlock/shape/try/shapes/triangles/'):
    triangles.append(cv2.imread(r'D:/sherlock/shape/try/shapes/triangles/{0}'.format(filepath),0))
#print(triangles)

for i in range(0,100):
    triangles[i]=triangles[i].reshape(1,784)
circles = []

# Load in the images
for filepath in os.listdir(r'D:/sherlock/shape/try/shapes/circles/'):
    circles.append(cv2.imread(r'D:/sherlock/shape/try/shapes/circles/{0}'.format(filepath),0))

for i in range(0,100):
    circles[i]=circles[i].reshape(1,784)
for i in range(0,100):
    triangles.append(circles[i])
    
input_=triangles

c=[0,1]
f=[]
for i in range(0,100):
    f.append(c)
t=[1,0]
for i in range(0,100):
    f.append(t)
from random import shuffle 
y=[]
for i in range(0,200):
    y.append(i)
print (y)
shuffle(y)
print(y)
ind=[]
for i in range(0,200):
    k=[y[i], f[i], input_[i]]
    ind.append(k)
print(ind[0])
ind.sort()
print(ind[0])
f1=[]
input_1=[]
for i in range(0,200):
    input_1.append(ind[i][2])
    f1.append(ind[i][1])
    
input_1=np.array(input_1).astype('float32')/255
f1=np.array(f1)
input_1=input_1.reshape(200,784)
print(input_1.shape)
```

```python
from keras.layers import Input, Dense
from keras.models import Model
import numpy as np

import matplotlib.pyplot as plt

input_data= Input(shape=(784,))
l1= Dense(units=100, activation='relu')(input_data)
l2 = Dense(units=50, activation='relu')(l1)
l3 = Dense(units=10, activation='relu')(l2)
l4 = Dense(units=2, activation='softmax')(l3)
net=Model(input_data,l4)

sgd = SGD(lr=0.1, momentum=0.9)
net.compile(optimizer='sgd', loss='binary_crossentropy', metrics=['accuracy'])
net.fit(input_1,f1,epochs=1500)


```

```python
import matplotlib.pyplot as plt

p=input_1[125]
p=p.reshape(1,784)

re=net.predict(p)
print(re)
if re[0][1]>=0.5:
    print('triangle')
else:
    print ('circle')
p=p.reshape(28,28)
plt.imshow(p)
plt.show() 
```
